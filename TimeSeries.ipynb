```
ticker = "INTC" #@param {type:"string"}
api_key = "83BUA0TX3AHFD911" #@param {type:"string"}
duration = "DAILY" #@param ["DAILY", "WEEKLY"]
duration_dict={"DAILY": "TIME_SERIES_DAILY_ADJUSTED", "WEEKLY": "TIME_SERIES_WEEKLY_ADJUSTED"}
```

```
import json
import requests

response = requests.get("https://www.alphavantage.co/query?function=" + duration_dict[duration] + "&symbol="+ ticker +"&outputsize=full&apikey=" + api_key)

if response.status_code == 200:
  stock_data = json.loads(response.content.decode('utf-8'))
else:
  print("Fail to get stock data")
```

```
if duration == "DAILY":
  duration_field = "Time Series (Daily)"
else:
  duration_field = 'Weekly Adjusted Time Series'
for date in stock_data[duration_field]:
  print(date)
  for col in stock_data[duration_field][date]:
    print("%s %s" % (col,stock_data[duration_field][date][col]))
  break
  
```
